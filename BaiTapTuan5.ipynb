```python
#W5A3
def prime(a):
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            print(a,"ko la so nguyen to")
            break
    else:
        print(a,"la so nguyen to")
a = int(input())
prime(a)
```

```python
#W5A4
#so hoan hao la so co tong cac uoc(tru no) bang no
def perfect(a):
    tong = 0 
    for i in range(1,a):
        if a % i == 0:
            tong += i 
    if tong == a:
        print(a,"la so hoan hao")
    else:
        print(a,"ko la so hoan hao")
a = int(input())
perfect(a)
```

```python
#W5A5
def giatri(A,k):
    for i in range(len(A)):
        if A[i]==k:
            print(i)
            break
    else:
        print("-1")
A = list(map(int,input().split()))
k = int(input())
giatri(A,k)
        
```

```python
#W5A6
def factorial(n):
    tich = 1
    for i in range(1,n+1):
        tich *= i
    return tich
n = int(input())
factorial(n)
```

```python
#W5A7
a,b = map(int,input().split(' '))
pt = input()
def mt(a,b,pt):
    if pt == "-":
        return a-b
    if pt == "+":
        return a+b
    if pt == "*":
        return a*b
    if pt == "/":
        return a/b
mt(a,b,pt)

    
```

```python
#W5A8
def bit(n):
    ketqua = str()
    while n > 0:
        ketqua = str(n%2) + ketqua
        n//=2
    return ketqua
def kc_Hamming(a,b):
    count = 0 
    vi_tri_khac_nhau = a ^ b
    for i in bit(vi_tri_khac_nhau):
        if i == "1":
            count+=1
    return count
a,b = map(int,input().split(' '))
kc_Hamming(a,b)
```

```python
#W5A9
def tong_chu_so(n):
    sum = 0
    while n > 0:
        sum += n%10
        n//=10
    return sum
n = int(input())
tong_chu_so(n)
```

```python
#W5A10
a,b = map(str,input().split(' '))
def anh_xa(n):
    return chr(ord('z')-(ord(n)-ord('a')))
def dang_cau(a,b):
    for i in range(len(a)):
        if a[i]==anh_xa(b[i]) or a[i]==b[i]:
            continue
        else:
            return False
    return True
dang_cau(a,b)



```
