```python
#W6A4
input_str = input("nhap cac cap key:value cách nhau bởi dấu cách")
pairs = input_str.split()
result_dict = {}
for pair in pairs:
    if ':' in pair:
        key, value = pair.split(':')
        if key not in result_dict:
            result_dict[key]=[value]
        else:
            result_dict[key].append(value)
print(result_dict)
```

```python
#W6A9:
dict1 = eval(input("nhap dictionary 1"))
dict2 = eval(input("nhap dictionary 2"))

result_dict = dict1.copy()
for key, value in dict2.items():
    if key in result_dict:
        result_dict[key] += value
    else:
        result_dict[key] = value
    #sap xep
    sorted_dict = dict(sorted(result_dict.items()))
print(sorted_dict)

```

```python
#W6A10
A = list(map(int,input().split()))
k = int(input())
A.sort()
B=[]
for i in A:
    for j in A:
        if i + j == k and i<=j:
            pair = (i, j)
            if pair not in B:
                B.append(pair)
print(B)


```

```python
#W6A12
a = list(map(int,input().split()))
b = {}
for i in a:
    if i in b:
        b[i] += 1
    else:
        b[i] = 1
result_num = None
max_count = 0
for num,count in b.items():
    if count > max_count:
        max_count = count
        result_num = num
    elif count == max_count:
        if result_num is None or num < result_num:
            result_num = num
print(result_num)
 
```

```python
#W6A14
A = list(map(int,input().split()))
B = list(map(int,input().split()))
C = []
A = sorted(list(set(A)))
B = sorted(list(set(B)))
for i in A:
    for j in B:
        if i == j:
            C.append(i)
print(C)
```
