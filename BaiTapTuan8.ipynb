```python
#W7A19
def merge(left,right):
    res = []
    i=j=0
    while i<len(left) and j<len(right):
        if left[i] < right[j]:
            res.append(left[i])
            i+=1
        else:
            res.append(right[j])
            j+=1
    res.extend(left[i:])
    res.extend(right[j:])
    print(*res)
left = list(map(int,input().split()))
right = list(map(int,input().split()))
merge(left,right)
```

```python
#W7A27
s = input()
stack = []
mapping = {')': '(', ']': '[', '}': '{'}
is_valid = True

for char in s:
    if char in "([{":  
        stack.append(char)
    elif char in ")]}":  
        if not stack or stack[-1] != mapping[char]:
            is_valid = False
            break
        stack.pop()  
if is_valid and len(stack) == 0:
    print("YES")
else:
    print("NO")
```

```python
#W7A28
def hwei(n):
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:
            for j in range(i*i, n + 1, i):
                is_prime[j] = False     
    max_len = 0
    best_start = -1
    current_len = 0
    current_start = -1
    for i in range(4, n):
        if not is_prime[i]: 
            if current_len == 0:
                current_start = i
            current_len += 1
        else:
            
            if current_len > max_len:
                max_len = current_len
                best_start = current_start
            current_len = 0
    if current_len > max_len:
        max_len = current_len
        best_start = current_start
    if best_start != -1:
        return list(range(best_start, best_start + max_len))
    return []
n = int(input())
print(hwei(n))
```

```python
#W7A30
arr = list(map(int,input().split()))
n = len(arr)
L = [1] * n
P = [-1] * n
for i in range(1, n):
    for j in range(0, i):
        if arr[i] >= arr[j] and L[i] < L[j] + 1:
            L[i] = L[j] + 1
            P[i] = j
max_len = max(L)
pos = 0
for i in range(n):
    if L[i] == max_len:
            pos = i
            break
result = []
while pos != -1:
    result.append(arr[pos])
    pos = P[pos]
result.reverse()
print(*(result))
```

```python
#W7A31
def hwei(number_str):
    s = list(number_str)
    n = len(s)
    for i in range(n - 1):
        suffix = s[i+1:]
        min_val = min(suffix)
        if min_val < s[i]:
            last_idx_in_suffix = -1
            for k in range(len(suffix) - 1, -1, -1):
                if suffix[k] == min_val:
                    last_idx_in_suffix = k
                    break
            swap_idx = i + 1 + last_idx_in_suffix
            s[i], s[swap_idx] = s[swap_idx], s[i]
            result = "".join(s)
            return str(int(result))   
    s[n-1], s[n-2] = s[n-2], s[n-1]
    return "".join(s)
number_str = input()
print(hwei(number_str))
```
