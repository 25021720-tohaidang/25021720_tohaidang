```python
#W7A1
def binary_search(arr,target):
    arr.sort()
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
arr = list(map(int,input().split(' ')))
target = int(input())
print(binary_search(arr,target))
```

```python
#W7A1
def binary_search(a,l,r,x):
    mid=(l+r)//2
    if l==r:
        if a[mid]==x:
            return mid+1
        return -1
    elif a[mid]<x:
        return binary_search(a,mid+1,r,x)
    else:
        return binary_search(a,l,mid-1,x)
arr = list(map(int,input().split()))
n = int(input())
print(binary_search(arr,0,len(arr)-1,n))
```

```python
#W7A3
#Bubble Sort
def Bubble_Sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
                swapped = True
        if not swapped:
            break
    return arr


#Selection Sort
def Selection_Sort(arr):
    n = len(arr)
    for i in range(n):
        min_idex = i
        for j in range(i+1,n):
            if arr[j] < arr[min_idex]:
                min_idex = j
        arr[i],arr[min_idex] = arr[min_idex],arr[i]
    return arr

arr = list(map(int,input().split(' ')))
print(Bubble_Sort(arr))
print(Selection_Sort(arr))


```

```python
#W7A4
def find(arr):
    count = {}
    for num in arr:
        if num in count:
            count[num] += 1
        else:
            count[num] = 1
    max_count = 0
    result_num = None
    checked = set()
    for num in arr:
        if num in checked:
            continue
        checked.add(num)
        if count[num] > max_count:
            max_count = count[num]
            result_num = num
    print(f'{result_num} xuat hien nhieu nhat,som nhat,{max_count} lan')
arr = list(map(int,input().split(' ')))
find(arr)

```

```python
#W7A7
def length_of_lis(nums):
    if not nums:
        return 0
    dp = [1] * len(nums)
    
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
                
    return max(dp)
nums = list(map(int,input().split()))
print(length_of_lis(nums))
```

```python
#W7A13
arr = list(map(int,input().split()))
arr.sort()
max_count = 1
current_count = 1
for i in range(len(arr)):
    for j in range(i+1,len(arr)):
        if (arr[j] - arr[i] == 1) or (arr[i] == arr[j]):
            current_count+=1
            if current_count > max_count:
                max_count = current_count
        else:
            current_count = 1
print(max_count)
```
