```python
#W9A1
lst1 = list(map(int,input().split()))
lst2 = list(map(int,input().split()))
def merge(a,b):
    i=j=0
    arr = []
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            arr.append(a[i])
            i+=1
        else:
            arr.append(b[j])
            j+=1
    arr.extend(a[i:])
    arr.extend(b[j:])
    print(*arr)
merge(lst1,lst2)
```

```python
#W9A3
def smallest(n):
    s = list(n)
    length = len(s)
    for i in range(length):
        min_digit = s[i]
        pos = -1
        for j in range(i + 1, length):
            if s[j] < min_digit:
                min_digit = s[j]
                pos = j
        if pos != -1:    
            if i == 0 and s[pos] == '0':
                continue
            s[i], s[pos] = s[pos], s[i]
            break
    return ''.join(s)
n = input()
smallest(n)

```

```python
#W9A4
def can_transform(a, b):
    i = 0  
    for ch in a:
        if i < len(b) and ch.upper() == b[i]:   
            i += 1
        elif ch.islower():  
            continue
        else:   
            return "NO"
    return "YES" if i == len(b) else "NO"
a = input()
b = input()
print(can_transform(a,b))
```

```python
#W9A6
def hoan_vi_tiep_theo(a,b):
    i = a - 2
    while  i >= 0 and b[i]>=b[i+1]:
        i-=1
    if i>=0:
        j = a - 1
        b[i],b[j] = b[j],b[i]
    left = i+1
    right = a-1
    while left<right:
        b[left],b[right] = b[right],b[left]
        left+=1
        right-=1
    print(*b)
n = int(input())
lst = list(map(int,input().split()))
hoan_vi_tiep_theo(n,lst)
```

```python
#W9A7
flowerbed = list(map(int,input().split()))
k = int(input())
def canPlaceFlowers(flowerbed,k):
    for i in range(len(flowerbed)):
        if flowerbed[i] == 0:
            left = (i==0) or (flowerbed[i-1] == 0)
            right = (i == len(flowerbed)-1) or (flowerbed[i+1] == 0)
            if left and right : 
                flowerbed[i] == 1
                k -= 1
    return k <= 0
print(canPlaceFlowers(flowerbed,k))

```

```python
#W9A8
def  is_path_crossing(moves):
    x,y = 0,0
    visited = {(0,0)}
    for i in moves:
        if i == 'R':
            x += 1
        elif i == 'L':
            x -= 1
        elif i == 'U':
            y += 1
        elif i == 'D':
            y -= 1
        if (x,y) in visited:
            return True
        visited.add((x,y))
    return False
moves = input()
print(is_path_crossing(moves))


```

```python
#W9A9
def Median(a,b,c,d,e):
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > e:
        b, e = e, b
    if b > c:
        b, c = c, b
    return b
a,b,c,d,e = map(int,input().split(' '))
Median(a,b,c,d,e)
```

```python
#W9A10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    i = 0
    res = 0
    for house in houses:
        while i < len(heaters) - 1 and abs(heaters[i + 1] - house) <= abs(heaters[i] - house):
            i += 1
        res = max(res, abs(heaters[i] - house))
    return res
houses = list(map(int,input().split()))
heaters = list(map(int,input().split()))
findRadius(houses,heaters)
```
