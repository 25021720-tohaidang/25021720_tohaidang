```python
#W9A1
lst1 = list(map(int,input().split()))
lst2 = list(map(int,input().split()))
def merge(a,b):
    i=j=0
    arr = []
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            arr.append(a[i])
            i+=1
        else:
            arr.append(b[j])
            j+=1
    arr.extend(a[i:])
    arr.extend(b[j:])
    print(*arr)
merge(lst1,lst2)
```

```python
#W9A5
lst1 = list(map(int,input().split()))
lst2 = list(map(int,input().split()))
def check(a,b):
    a1 = " ".join(map(str,a))
    b1 = " ".join(map(str,b))
    if a1 in b1:
        return "YES"
    return "NO"
print(check(lst1,lst2))

```

```python
#W9A6
def hoan_vi_tiep_theo(a,b):
    i = a - 2
    while  i >= 0 and b[i]>=b[i+1]:
        i-=1
    if i>=0:
        j = a - 1
        b[i],b[j] = b[j],b[i]
    left = i+1
    right = a-1
    while left<right:
        b[left],b[right] = b[right],b[left]
        left+=1
        right-=1
    print(*b)
n = int(input())
lst = list(map(int,input().split()))
hoan_vi_tiep_theo(n,lst)
```

```python
#W9A7
flowerbed = list(map(int,input().split()))
k = int(input())
def canPlaceFlowers(flowerbed,k):
    for i in range(len(flowerbed)):
        if flowerbed[i] == 0:
            left = (i==0) or (flowerbed[i-1] == 0)
            right = (i == len(flowerbed)-1) or (flowerbed[i+1] == 0)
            if left and right : 
                flowerbed[i] == 1
                k -= 1
    return k <= 0
print(canPlaceFlowers(flowerbed,k))

```

```python
#W9A8
def  is_path_crossing(moves):
    x,y = 0,0
    visited = {(0,0)}
    for i in moves:
        if i == 'R':
            x += 1
        elif i == 'L':
            x -= 1
        elif i == 'U':
            y += 1
        elif i == 'D':
            y -= 1
        if (x,y) in visited:
            return True
        visited.add((x,y))
    return False
moves = input()
print(is_path_crossing(moves))


```
