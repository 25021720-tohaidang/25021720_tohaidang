```python
#W6A16
n,m = map(int,input().split(' '))
matrix = []
print(f'nhap {n} dong moi dong la {m} so nguyen')
for i in range(n):
    row = list(map(int,input().split()))
    matrix.append(row)
print("ma tran dinh dang can phai 4 vi tri")
for row in matrix:
    for val in row:
        print(f'{val:>4}',end='')
    print()
```

```python
#W6A17
n = int(input())
matrix = []
for _ in range(n):
    row = list(map(int,input().split()))
    matrix.append(row)
print("duong cheo chinh cua ma tran la")
for i in range(n):
    print(matrix[i][i],end=' ')
print()
print("duong cheo phu cua ma tran la")
for i in range(n):
    print(matrix[i][n-i-1],end=' ')
```

```python
#W6A18
n,m,k = map(int,input().split(' '))
matrix = []
if k <= (n-1) and k <= (m-1):
    print(f"nhap {n} dong moi dong la {m} so nguyen")
    for _ in range(n):
        row = list(map(int,input().split()))
        matrix.append(row)
    print(f'tong cac phan tu tren cot {k} la')
    sum1 = 0
    for i in range(n):
        sum1 += matrix[i][k-1]
    print(sum1)
    print()
    print(f'tong cac phan tu o hang {k} la')
    sum2 = 0
    for j in range(m):
        sum2 += matrix[k-1][j]
    print(sum2)
else:
    print(f"{k} ngoai khoang ma tran")
```

```python
#W6A20
n = list(map(int,input().split()))
k = int(input())
def gia_tri(a,b):
    try:
        return a.index(b)
    except ValueError:
        return -1
print(gia_tri(n,k))
```

```python
#W6A21
chuoi = input().split()
result = {}
#enumerate de lay ca vi tri(index) va gia tri (word)
for index, word in enumerate(chuoi): 
    if word not in result:
        result[word] = [index]
    else:
        result[word].append(index)
for key in result:
    result[key] = tuple(result[key])
print(result)
```

```python
#W6A26
so = list(map(int,input().split()))
tong_cuc_dai_dp = 0
for i in range(1,len(so)-1):
    if so[i] > so[i-1] and so[i] > so[i+1]:
        tong_cuc_dai_dp += so[i]
print(tong_cuc_dai_dp)
```
