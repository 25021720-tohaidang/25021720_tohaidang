```python
#W11A1
from pathlib import Path
WORKDIR = Path(".")
path = "W11A1.txt"
def reverse(path: str) -> int:
    myfile = WORKDIR/path
    with open(myfile, "r") as f:
        n = int(f.read().strip())
    n_rev = 0
    while n :
        n_rev = n_rev*10 + n%10
        n//=10
    return n_rev
print(reverse(path))
```

```python
#W11A4
from pathlib import Path
WORKDIR = Path(".")
path = "W11A4.txt"
def areAnagrams(path:str) -> bool:
    file = WORKDIR/path
    with open (file,"r") as f:
        lines = f.readlines()
        if len(lines) < 2:
            return False
        s1 = lines[0].strip()
        s2 = lines[1].strip()
        clean_s1 = sorted(s1.replace(" ", "").lower())
        clean_s2 = sorted(s2.replace(" ", "").lower())
        return clean_s1 == clean_s2
areAnagrams(path)

    
```

```python
#W11A8
from pathlib import Path
WORKDIR = Path(".")
path = "output.txt"
def fibonancci_words(n:int):
    if n < 0:
        return 
    words = [""] * (n+1)
    words[0] = "a"
    if n >= 1:
        words[1] = "b"
    for i in range(2,n+1):
        words[i] = words[i-1] + words[i-2]
    file = WORKDIR/path
    with open(file,"w") as f:
        for i in range(n+1):
            f.write(f'f({i}) = {words[i]}\n')
fibonancci_words(10)
```

```python
#W11A12
from pathlib import Path
WORKDIR = Path(".")
path = "W11A12.txt"
def maximumProduct(path: str) -> int:
    file = WORKDIR/path
    with open(file, 'r') as f:
        lines = f.readlines()
        nums = [int(x) for x in lines[1].split()]
    nums.sort()
    return max(nums[-1] * nums[-2] * nums[-3], nums[0] * nums[1] * nums[-1])
maximumProduct(path)
```

```python
#W11A13
from pathlib import Path
WORKDIR = Path(".")
path = "W11A13.txt"
def moveZeroes(path: str) -> list:
    file = WORKDIR/path
    with open(file, 'r', ) as f:
        lines = f.readlines()
        if len(lines) < 2: return []
        nums = [int(x) for x in lines[1].split()]
    non_zero = [x for x in nums if x != 0]
    zeros = [0] * (len(nums) - len(non_zero))
    return non_zero + zeros
moveZeroes(path)  
```

```python
#W11A14
from pathlib import Path
WORKDIR = Path(".")
path = "W11A14.txt"
def averageTime(path: str):
        file = WORKDIR/path
        total_time = 0.0
        count = 0
        with open(file, 'r',) as f:
            for line in f:
                parts = line.strip().split(',')
                if len(parts) >= 2:
                    total_time += float(parts[1])
                    count += 1       
        if count == 0:
            print(0.00)
        else:
            avg = total_time / count
            print(f"{avg:.2f}")
averageTime(path)

   
```

```python
from pathlib import Path
WORKDIR = Path(".")
path = "W11A15.txt"
def findMovies (path: str) -> list:
    file = WORKDIR/path
    res = []
    with open(file,"r") as f:
        lines = f.readlines()
        for line in lines:
            chuoi = line.strip().split(',')
            if len(chuoi) < 4:
                continue
            movie_name = chuoi[1].strip()
            description = chuoi[2].strip()
            rating = float(chuoi[3].strip())
            if rating > 8.0 and 'boring' not in description:
                res.append(movie_name)
    if not res:
        return "All are boring!"
    return res
findMovies(path)

```

```python
#W11A16
from pathlib import Path
WORKDIR = Path(".")
path = "W11A16.txt"
import math
def calculate_perimeter(path: str):
    file = WORKDIR/path
    with open(file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
            
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if not line: 
            i += 1
            continue
                
        if line.startswith("TYPE: SQUARE"):
            param_line = lines[i+1].strip()
            side = float(param_line.split(':')[1])
            perimeter = 4 * side
            print(f"{perimeter:.2f}")
            i += 2
                
        elif line.startswith("TYPE: RECTANGLE"):
            param_line = lines[i+1].strip()
            parts = param_line.split()
            width = float(parts[1])
            height = float(parts[3])
            perimeter = 2 * (width + height)
            print(f"{perimeter:.2f}")
            i += 2
            
        elif line.startswith("TYPE: CIRCLE"):
            param_line = lines[i+1].strip()
            radius = float(param_line.split(':')[1])
            perimeter = 2 * math.pi * radius
            print(f"{perimeter:.2f}")
            i += 2
        else:
            i += 1
calculate_perimeter(path)
```

```python
#W11A17
from pathlib import Path
WORKDIR = Path(".")
path = "W11A17.txt"
def productExceptSelf (path: str) -> list[int]:
    file = WORKDIR/path
    res = []
    with open(file,"r") as f:
        lines = f.readlines()
        line = lines[1].strip().split()
        tich = 1
        for i in line:
            tich *= int(i)
        for j in line:
            res.append(int(tich/int(j)))
    print(*res)
productExceptSelf(path)
```

```python
#W11A18
from pathlib import Path
WORKDIR = Path(".")
path = "W11A18.txt"
import math
def estimatedTime(path: str):
    file = WORKDIR/path
    with open(file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        src_parts = lines[0].strip().split()
        x1 = float(src_parts[1].split(':')[1])
        y1 = float(src_parts[2].split(':')[1])

        des_parts = lines[1].strip().split()
        x2 = float(des_parts[1].split(':')[1])
        y2 = float(des_parts[2].split(':')[1])
        
        vel_line = lines[2].strip()
        vel_str = vel_line.split(':')[1].split()[0]
        velocity = float(vel_str)
            
        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        time = distance / velocity
        
        print(f"{time:.2f}")
estimatedTime(path)
```
